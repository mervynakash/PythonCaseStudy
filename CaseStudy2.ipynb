```python
import pandas  as pd
import numpy as np
import os
import math
import datetime

os.chdir("E:/Python DS/Assignments")

store = pd.read_csv("DataSets/Store.csv", encoding='latin')
```

```python
print(store.shape)
print(store.columns)
```

```python
# How many unique cities are the orders being delivered to
print(store.City.unique())
```

```python
# What is the total quantity sold in the East Region?
print(store[store.Region == "East"]['Quantity'].count())
print(store.Quantity[store.Region == 'East'].count())
```

```python
# Find the sum of the quantity sold in the East Region
print(store.Quantity[store.Region == 'East'].sum())
```

```python
# In the south region sort the sales in decreasing order
print(store[store.Region == 'South'].sort_values(by = 'Sales'))
```

```python
# Find the mean of quantity for every region
print(store.groupby(by = 'Region').mean()['Quantity'])
```

```python
# Find the mean of sales for every category
print(store.groupby(by = 'Category').mean()['Sales'])
```

```python
# Find the max, min, sum of sales and profit for every category
grp_cat = store.groupby(by = 'Category')
for i in ['Sales','Profit']:
    print("For",i,":")
    print(grp_cat.min()[i])
    print()
    print(grp_cat.max()[i])
    print()
    print(grp_cat.sum()[i])
    print("================================")
```

```python
# Find the max, min, sum of sales and profit for every category
store.groupby('Category')[['Sales','Profit']].agg(['min','max','sum'])
```

```python
# Find sum of sales and max profit for every segment
grp_seg = store.groupby(by = 'Segment')
print(grp_seg.sum()['Sales'])
print(grp_seg.max()['Profit'])
```

```python
store.groupby('Segment').agg({'Sales':np.sum, 'Profit':np.max})
```

```python
# For every segment find the mean of the discount
print(grp_seg.mean()['Discount'])
```

```python
# For every segment find the most profitable customers
max_pro = store.groupby(by='Segment').max()['Profit']
type(max_pro)
print("For Consumer Segment the max profit has been made by ", store[store.Profit==max_pro.Consumer]['Customer Name'].values)
print("For Corporate Segment the max profit has been made by ", store[store.Profit==max_pro.Corporate]['Customer Name'].values)
print("For Corporate Segment the max profit has been made by ", store[store.Profit==max_pro['Home Office']]['Customer Name'].values)
```

```python
# What are the top 5 categories that give maximum profit?
cat_profit = (store.groupby(by = 'Sub-Category').max()['Profit'])
print(cat_profit.sort_values(ascending=False).head())
```

```python
# What is the Total Sales, Quantity, Discount, Profit across Total US.
print("Total Sales:",store.Sales.sum())
print("Total Quantity:",store.Quantity.sum())
print("Total Discount:",store.Discount.sum())
print("Total Profit:",store.Profit.sum())
```

```python
# How many times has it taken more than 5 days from placing an order to shipping
store['Order Date'] = pd.to_datetime(store['Order Date'])
store['Ship Date'] = pd.to_datetime(store['Ship Date'])
noday = store['Ship Date'] - store['Order Date']
noday = (noday / np.timedelta64(1,'D')).astype(int)
count = 0
for i in noday:
    if(i > 5):
        count = count + 1
    else:
        continue
print("The number of days it took to place an order to shipping:",count)
```

```python
# Find the total number of orders in every category which has been shipped with a duration > 5 days
pos = list()
pos2 = list()
for i in range(len(noday)):
    if(noday[i] > 5):
        pos.append(i)
    else:
        pos2.append(i)
storenew = store.iloc[pos,:]
countcat5 = storenew.groupby(by='Category').count()['Order ID']
print(countcat5)
```

```python
# What’s the percentage of items which has been shipped within 5 days
storenew2 = store.iloc[pos2,:]
countcat = storenew2.groupby(by='Category').count()['Order ID']
percless5 = countcat*100/storenew2.shape[0]
print(percless5)
```

```python
# What’s the percentage of items which has been shipped after 5 days
percmore5 = countcat5*100/storenew.shape[0]
print(percmore5)
```
